# Pipeline for app

We will first def a function to extract the predicted words. Then we will def a function to delete from the predicted words those who are not in our pictograms. Finally, we will def a function to get the k most likely words. The pipeline will be model_output -> extract_predicted_words() -> filter_according_to_pictograms() -> get_top_k_predictions()

Def extract_predicted_words

```python
"this can be improved with a list comprehension"
predicted_words = []
def extract_predicted_words(unmasked, predicted_words):
  for unmask in unmasked: 
    predicted_words.append(unmask['token_str'])
  return predicted_words


predicted_words = extract_predicted_words(unmasked, predicted_words)

predicted_words, len(predicted_words)

```

Def filter_according_to_pictograms()

```python

def filter_according_to_pictograms(predicted_words): 
  predicted_words[:] = [word for word in predicted_words if word in pictograms]
  return predicted_words

filtered_pictograms = filter_according_to_pictograms(predicted_words)

print(filtered_pictograms)

```

def get_top_k_predictions 

```python


def get_top_k_predictions(filtered_pictograms, k):
  if len(filtered_pictograms) > k:
    top_k_predictions = dict(zip(filtered_pictograms[:k], [unmasked[score]['score'] for score in range(len(filtered_pictograms[:k]))]))
  else: 
    top_k_predictions = dict(zip(filtered_pictograms, [unmasked[score]['score'] for score in range(len(filtered_pictograms))]))
  return top_k_predictions

top_k_predictions = get_top_k_predictions(filtered_pictograms, 8)

top_k_predictions

```

We wrap the whole pipeline in an app (i.e., def fill_mask())

```python
import gradio as gr
from transformers import pipeline

top_k = 1000 #@param {type: "integer"}

unmasker = pipeline("fill-mask", model="CenIA/distillbert-base-spanish-uncased")

def fill_mask(text): 
  text += ' [MASK]'
  unmasked = unmasker(text, top_k=top_k)
  predicted_words = []
  predicted_words = extract_predicted_words(unmasked, predicted_words)
  filtered_pictograms = filter_according_to_pictograms(predicted_words)
  top_k_predictions = get_top_k_predictions(filtered_pictograms, 8)
  return top_k_predictions

examples = [["quiero comer "], ["me gustaría ir a"], ["yo"]]

demo = gr.Interface(fn=fill_mask, inputs=gr.Textbox(lines=1, label="Input"), outputs=gr.Label(label="Output"), examples=examples)
demo.launch(debug=True, share=True)
```

# Using contexts

For this part, we will filter the predicted words according to another variable, i.e., the 'context' that is provided, which in turn will filter the output according to a list of pictograms contained in a given 'context'

First, we will download the contexts from a github repo, and inspect it a little bit. 

```python
import os.path
from os import path

# download 'contextos' file
if not path.exists("contextos2.json"):
  !wget "https://raw.githubusercontent.com/franfram/Transformers-nlp/main/contextos2.json"


# open 'contextos' file
import json
with open("contextos2.json") as archivo: 
  contextos = json.load(archivo)

```

Now, all the pictograms are contained within 'contexts'

```python
contextos
```

And this 'contextos' object is a dictionary containing 10 context keys, which in turn contain a list of the given pictograms of a context. 

```python
contextos.keys(), len(contextos.keys()), type(contextos['desayuno'])


```

Since 'contextos' contains compound words, we will have to wrangle the data a little bit so that we don't have compound words anymore. 


```python
import re

# def split_contextos(contextos): 
#   return {k: ' '.join(contextos[k]).split(' ') for k, v in contextos.items()}

  

def split_contextos(contextos): 
  return {k: set(re.findall(r"[\w']+|[.,!?;]", ' '.join(contextos[k]))) for k, v in contextos.items()}



contextos_split = split_contextos(contextos)
contextos_split


```

We will modify the `filter_according_to_pictograms` function to `filter_according_to_contexts`. To accomplish this, instead of filtering the output to match whole set of pictograms, we will filter the output to match the subset of pictograms contained in a given context (which will be another function argument)


We first extract the output of the model: 

Now we create the `filter_according_to_contexts()` function


```python
def filter_according_to_context(predicted_words, context): 
  predicted_words[:] = [word for word in predicted_words if word in contextos_split[context]]
  return predicted_words

filtered_pictograms = filter_according_to_context(predicted_words, context = 'merienda')



```
