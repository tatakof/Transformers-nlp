# Text generation pipeline

Flax community, GPT-2 spanish

```python
from transformers import pipeline


generator = pipeline("text-generation", model="flax-community/gpt-2-spanish")

prompt = "quiero comer un  "#@param {type: "string"}

n_words_prompt = len(prompt.split())
max_length = 5 #@param {type: "integer"}
# Add `n_words_prompt` to avoid having problems due to `max_length` being smaller
# than the amount of words in the prompt
max_length += n_words_prompt

num_return_sequences = 30 #@param {type: "integer"}


# generate text
generator(
    prompt, 
    max_length=max_length,
    num_return_sequences=num_return_sequences,
)
```

# Mask filling pipeline (fill in the blanks in a given text, 1-gram prediction)

# model="dccuchile/bert-base-spanish-wwm-uncased"

```python
from transformers import pipeline
top_k = 8 #@param {type: "integer"}

unmasker = pipeline("fill-mask", model="dccuchile/bert-base-spanish-wwm-uncased")
text = 'quiero comer un ' #@param {type: "string"}
text += ' [MASK]'

result = unmasker(text, top_k=top_k); result
```

## Extraemos las predicciones

```python
print([r["token_str"] for r in result])
```
